```python
#Most frequent actors:
top_actors = merged_data[merged_data["role"] == "ACTOR"]["name"].value_counts().head(10)
top_actors.plot(kind="barh", title="Top 10 Actors")
plt.show()
```

```python
top_directors = merged_data[merged_data["role"] == "DIRECTOR"]["name"].value_counts().head(10)
top_directors.plot(kind="barh", title="Top 10 Directors")
plt.show()
```

```python
# Detecting outliers in imDb_score
min, max = df_titles.imdb_score.min(), df_titles.imdb_score.max()
min,max
```

```python
avg, std = df_titles.imdb_score.mean(), df_titles.imdb_score.std()
avg, std
```

```python
#here 3-std deviation method will be used to find outlier
higher_limit = avg + 3*std
higher_limit 
```

```python
#if max value is higher than higher_limit, then max value is the outlier, here the max value is 9.9 which is lower than 
#higher limit 10.0079, so there is no outlier present in imdb_score.
```
